```python
race_count = df.value_counts("race")
race_count
```

```python
mask = df["sex"] == "Male"
mean_male_age = df.loc[mask, ["age"]].mean() # series, drop square brackets on "age" to make it a float
print(mean_male_age)
print(type(mean_male_age))
```

```python
num_records = df["education"].value_counts().sum()
print(df["education"].value_counts())
print(len(df))
```

```python
mask = df["education"] == "Bachelors"
float(df.loc[mask, "education"].count() / num_records * 100)
```

```python
filter_vals = ["Bachelors", "Masters", "Doctorate"]
advanced_ed_df = df.loc[~df["education"].isin(filter_vals)]
advanced_ed_df.head()
```

```python
higher_education = df.loc[df["education"].isin(["Bachelors", "Masters", "Doctorate"])]
lower_education = df.loc[~df["education"].isin(["Bachelors", "Master", "Doctorate"])]

higher_education_rich = higher_education.loc[higher_education["salary"] == ">50K", "salary"].count() / df.shape[0]
lower_education_rich = lower_education.loc[lower_education["salary"] == ">50K", "salary"].count() / df.shape[0]

print(higher_education_rich)
print(lower_education_rich)
```
